```python
#!/usr/bin/env python

# ------------------------------------------------------------------------------------------------------------------- #
# Import Required Libraries
# ------------------------------------------------------------------------------------------------------------------- #
import os
import re
import glob
import numpy as np
import pandas as pd
from datetime import date
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, FixedLocator
import math
# Paths Of Directories
# ------------------------------------------------------------------------------------------------------------------- #
DIR_HOME1 = 'D:\\Work\\OneDrive\\Walop_work\\Exptime_cal\\'
# ------------------------------------------------------------------------------------------------------------------- #

file_name1 = 'observed_pol.csv'
data_df = pd.read_csv(DIR_HOME1 + file_name1)

data_df
```

```python
# R band zero magnitude star flux and photons #
Mflux=4246   #corrections done for the SDSS filter bands# 
lc=6152      #central wavelength#
dl=0.221      #bandwidth/lc#
Ph=Mflux*1.51*10**7*(dl)/(10000*dl*lc) #photons/s/cm^2/angstrom#
Ph
```

```python
#Telescope and instrument throughput#
Pdia=129          #Diamter of the primary mirror#
Sdia=55.5        #secondary mirror obscuration including the new Baffle size# 
Obs=((Pdia/2)**2-(Sdia/2)**2)/(Pdia/2)**2          #Obscuration ratio#
#Transmission parameters#
mir=0.9           #Telescope mirrors reflectivity#
Iwin=0.99         #instrument window#
len1=0.99         #Collimator lenses-6#
CalibHWP=1    
fil=1
Wedge=0.99
HWP=0.98
wol=0.92
DispCor=0.99         #Dispersion Corrector#
Mirror=0.99
PBS=0.9              #Polarizing Beam Splitter#
len2=0.99            #camera Lenses#
Dwin=0.99            #Dewar Window#
CCD=0.9              #QE of CCD#
etainst=Obs*(mir**2)*(len1**6)*(CalibHWP)*(fil**1)*Wedge*(HWP**1)*(wol**1)*(DispCor**2)*(Mirror**1)*(PBS**1)*(len2**7)*(Dwin**1)*(CCD**1)
CCDrdnoise=3    #photons#
CCDdknoise=0.05  #photons/second#
pixscale=0.51    #arcseconds/pixel#
etainst
```

```python
#Telescope efficiency#
airmass=data_df['airmass']
Ele=(math.pi/2-np.arccos(1/airmass))*180/math.pi            #Elevation#
OPL=1/np.sqrt(1-0.96*np.sin((90-Ele)*math.pi/180)**2)       #Optical Path Length#
Atmext=0.9 #Attmospheric Extinction at zenith#
area=math.pi*(Pdia/2)**2            #Collecting Area of the Telescope#
FilBW=1363                          #Filter Bandwidth#
Eff=etainst*Atmext*np.cos((90-Ele)*np.pi/180)    #Observing Efficiency#
```

```python
#Auto-guider information#
Skymag=20  #mag/arc^2#
Centwdw=10  #arcsecs#
SNR=10      
Agmag=15     #Magnitude of the star for autoguider#
Ageta=Eff      
Agph=Ph*Eff*10**(-0.4*Agmag)*FilBW*area   #Photons for guide star#
Skyphd=Ph*10**(-0.4*Skymag)              #photons/s/cm^2/Angstrom/arcsec^2#
Agskyph=Skyphd*(Centwdw)**2*area*FilBW*etainst   
t=5/60                                    #Minutes to reach the SNR#
SigX=np.sqrt(((Agph+Agskyph)*t*60)/(Agph*t*60)**2)     #Error in the centroding#
```

```python
#Sky photons calculations#
Seeing=1 #arcseconds#
Phap1=data_df['inner_aperture0']           
Phap2=data_df['inner_aperture2']
Effseeing=np.sqrt(Seeing**2+SigX**2)          #seeing correction#
Aperarea1=math.pi*(Effseeing*Phap1/2)**2     #arcseconds#
Aperarea2=math.pi*(Effseeing*Phap2/2)**2      
Aperareap1=Aperarea1/pixscale               #pixels Spots 1 and 2
Aperareap2=Aperarea2/pixscale              #pixels  Spots 3 and 4
Skyph1=Skyphd*Aperarea1                    #Sky photons#                 
Skyph2=Skyphd*Aperarea2 
Skymag1=-2.5*np.log10(Skyph1/Ph)        #Sky Magnitude corresponding to sky photons#  
Skymag2=-2.5*np.log10(Skyph2/Ph)
Ext=-2.5*np.log10(Atmext)         #Extinction-mag/airmass#
Skymagext1=Skymag1-2.5*np.log10((1-0.6)+0.6*OPL)+Ext*(OPL-1)         #Extinction corrected magnitude#
Skymagext2=Skymag2-2.5*np.log10((1-0.6)+0.6*OPL)+Ext*(OPL-1)
Skyphext1=Ph*10**(-0.4*Skymagext1)    #photons/s/cm^2/Angstrom#
Skyphext2=Ph*10**(-0.4*Skymagext2)   #photons/s/cm^2/Angstrom#
```

```python
#Source info and photons#
Srcmag=data_df['Rmag'].astype(float)         #Magnitude of the Source#
pinp=data_df['p'].astype(float)              #Polarization Observed#
sigmapinp=data_df['p_er'].astype(float)      #P_err observed#
data_df['theta'] = data_df['theta'].astype(float)  #PA observed
EVPA=data_df['theta']
Nsrct=Eff*Ph*10**(-0.4*Srcmag)*FilBW*area          #total number photons#
Nsrc0=((1-pinp)*Nsrct/4)+(pinp*Nsrct/2*np.cos(EVPA)**2)          
Nsrc1=((1-pinp)*Nsrct/4)+(pinp*Nsrct/2*np.sin(EVPA)**2)
Nsrc2=((1-pinp)*Nsrct/4)+(pinp*Nsrct/2*np.cos((EVPA+math.pi/4))**2)
Nsrc3=((1-pinp)*Nsrct/4)+(pinp*Nsrct/2*np.sin((EVPA+math.pi/4))**2)
Nsky1=Skyphext1*area*FilBW*etainst/4                #Sky photons for aperture 0#
Nsky2=Skyphext2*area*FilBW*etainst/4                #Sky photons for aperture 2#
```

```python
#Photometry and Polarimetry#
Exptime=data_df['exposure']/60 
NoofExp=1
TperExp=Exptime/NoofExp
N0=Nsrc0*TperExp*60*NoofExp          #Total Number of Source Photons#
N1=Nsrc1*TperExp*60*NoofExp
N2=Nsrc2*TperExp*60*NoofExp
N3=Nsrc3*TperExp*60*NoofExp
TNsky1=Nsky1*TperExp*60*NoofExp      #Total Nuumber of Sky photons#
TNsky2=Nsky2*TperExp*60*NoofExp
Dark=CCDdknoise*TperExp*60*NoofExp        
Rdnoise=CCDrdnoise*NoofExp
TRdnoise1=Rdnoise*Aperareap1
TRdnoise2=Rdnoise*Aperareap2
sigmaN0=np.sqrt(N0+TNsky1+(Dark*Aperareap1)+(Rdnoise*Aperareap1))       # P error#
sigmaN1=np.sqrt(N1+TNsky1+(Dark*Aperareap1)+(Rdnoise*Aperareap1))
sigmaN2=np.sqrt(N2+TNsky2+(Dark*Aperareap2)+(Rdnoise*Aperareap2))
sigmaN3=np.sqrt(N3+TNsky2+(Dark*Aperareap2)+(Rdnoise*Aperareap2))
q=(N0-N1)/(N0+N1)
u=(N2-N3)/(N2+N3)
pout=np.sqrt(q**2+u**2)
EVPAn=0.5*(np.arctan(u/q)*180/math.pi)
sigmaq=np.sqrt((4*(N1**2*sigmaN0**2+N0**2*sigmaN1**2)/(N0+N1)**4))
sigmau=np.sqrt((4*(N3**2*sigmaN2**2+N2**2*sigmaN3**2)/(N2+N3)**4))
sigmapout=np.sqrt((q**2*sigmaq**2+u**2*sigmau**2)/(q**2+u**2))
sigmaEVPA=0.5*(sigmapout/pout)*180/math.pi
PSNR=pout/sigmapout
```
